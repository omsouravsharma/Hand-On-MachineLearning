```python
# Import Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Load data 

data = pd.read_csv("data/lifesat.csv")
data
```

```python
X = data[["GDP per capita (USD)"]].values
y = data[["Life satisfaction"]].values
```

```python
X
```

```python
y
```

```python
# Visualize data 

plt.scatter(X,y)
plt.axis([23_500, 62_500, 4,9])
plt.xlabel("GDP Per Capita (USD)")
plt.ylabel("Life Satisfaction")
plt.grid("True")
plt.show()
```

```python
# Linear Model
model = LinearRegression()

# Train the model
model.fit(X,y)
```

```python
# Make prediction
X_new = [[45_655.2]] # new country
model.predict(X_new)
```
